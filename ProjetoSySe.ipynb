```python
import pyttsx3
import requests
from bs4 import BeautifulSoup
import sys


# --- Funções Auxiliares ---


def falar(texto):
  """Realiza a leitura por voz do texto."""
  engine = pyttsx3.init()
  engine.setProperty('voice', 'brazil')
  engine.say(texto)
  engine.runAndWait()


def buscar_informacoes(duvida):
  """Simula a busca de informações (substitua com sua lógica real)."""
  resultados_simulados = [
  {"problema": f"Possíveis ataques relacionados a {duvida}:", "solucao": "Verifique as configurações de segurança."},
  {"problema": f"Prevenção contra {duvida}:", "solucao": "Mantenha seu software atualizado."},
  ]
  return resultados_simulados


# --- Agentes do Chatbot ---


def agente_buscador():
  """Agente Buscador (Etapa 1)."""
  print("Olá, sou a CySe sua ChabBot de ajuda para assuntos sobre Cyber Segurança.")
  print("Escolha uma das opções abaixo:")
  print("A - Informações de Cyber Segurança")
  print("B - Ajuda para um problema específico")


  escolha = input("Digite A ou B: ").upper()


  if escolha == "A":
   agente_especialista1()
  elif escolha == "B":
   agente_indicador1()
  else:
   print("Opção inválida. Escolha A ou B.")
   agente_buscador() # Recomeça


def agente_especialista1():
  """Agente Especialista 1 (Etapa 2)."""
  informacoes = [
  {"problema": "Phishing (e-mails falsos):", "solucao": "Não clique em links suspeitos."},
  {"problema": "Malware (vírus):", "solucao": "Use antivírus."},
  {"problema": "Senhas Fracas:", "solucao": "Use senhas fortes."},
  {"problema": "DDoS:", "solucao": "Proteja sua rede."},
  {"problema": "Engenharia Social:", "solucao": "Desconfie de pedidos de info."},
  ]


  print("\nTipos Comuns de Ataques Cibernéticos:")
  for info in informacoes:
   print(f"- {info['problema']} {info['solucao']}")
   falar(info["problema"] + info["solucao"])  # Chamada para falar() adicionada


  input("\nPressione Enter para voltar ao início.")
  agente_buscador()


def agente_indicador1():
  """Agente Indicador 1 (Etapa 3)."""
  duvida = input("\nDigite aqui sua dúvida/tema: ")
  agente_especialista2(duvida)


def agente_especialista2(duvida):
  """Agente Especialista 2 (Etapa 4)."""
  resultados = buscar_informacoes(duvida)


  print(f"\nSua Dúvida: {duvida}")
  if resultados:
   print("\nResultados Encontrados:")
  for resultado in resultados:
   print(f"- {resultado['problema']} {resultado['solucao']}")
   falar(resultado["problema"] + resultado["solucao"]) # Chamada para falar() adicionada
  else:
   print("\nNenhum resultado encontrado.")


  print("\nEspero que você tenha gostado da minha ajuda.")
  print("Escolha A para mais ajuda ou B para problema resolvido.")
  escolha = input("Digite A ou B: ").upper()


  if escolha == "A":
   agente_buscador()
  elif escolha == "B":
   agente_indicador2()
  else:
   print("Opção inválida.")
   agente_especialista2(duvida)


def agente_indicador2():
  """Agente Indicador 2 (Etapa 5)."""
  print("\nFico feliz em poder ter ajudado. Obrigado e se precisar, me chame!")
  sys.exit()


# --- Iniciar o Chatbot ---


agente_buscador()
```
